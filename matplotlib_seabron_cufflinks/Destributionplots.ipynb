```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use Distribution plot to visualize the distribution of quantitative data.

let's Discuss some plots that allows to visualize the distribution of data set. 
These plots are :
        * displot/histplot
        

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

Distplot/histplot

```python
# sns.displot(df['total_bill'],bins = 20)
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20 , kde=True) 
#both are same   kde is a kernal denisity estimater
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20 , kde=True)
```

Jointplot Compre two type of data

```python
sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='scatter')
```

```python
# sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='kde') 
# sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='hist')
# sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='hex') 
sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='reg') 
# sns.jointplot(x = 'total_bill' , y = 'tip' , data=df ,kind='resid') 
```

pairplot

```python
# sns.pairplot(df,hue='sex',palette='rainbow') 
sns.pairplot(df,hue='size',palette='rainbow') 
```
