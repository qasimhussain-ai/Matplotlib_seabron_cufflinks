```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

Matplotlib Basic

```python
x = np.linspace(0,5,11)
x
y = x**3
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("X_axis")
plt.ylabel("Y_axix")
```

Subplots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

Object Oriented Way

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

```python
fig = plt.figure(figsize = (20,8),dpi = 100)
x1 = fig.add_axes([.1,.1,1,1])
x1.plot(x,y)
y1 = fig.add_axes([.3,.3,.8,.8])
y1.plot(x,y)
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 5))

# X-axis data
xaxis = np.arange(0, 5, 0.5)

# Plot several lines with different styles
for i in range(10):
    ax.plot(
        xaxis, 
        xaxis + i, 
        color='red' if i < 3 else 'green' if i < 6 else 'blue',
        linewidth=1 + i * 0.1,
        linestyle=['-', '--', '-.', ':'][i % 4]
    )

# Add a few highlighted lines with markers
ax.plot(xaxis, xaxis + 14, color='purple', lw=1, ls='-', marker='o', markersize=4)
ax.plot(xaxis, xaxis + 15, color='purple', lw=1, ls='-', marker='o',
        markersize=8, markerfacecolor='red')
ax.plot(xaxis, xaxis + 16, color='purple', lw=1, ls='-', marker='s',
        markersize=8, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')

# Customize grid and labels
ax.set_title("Multiple Line Styles and Markers")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
