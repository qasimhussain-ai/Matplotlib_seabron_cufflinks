```python
import seaborn as sns
import pandas as pd 
import cufflinks as cf 
from plotly.offline import iplot

```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import plotly.express as px
fig = px.line(tips, y='total_bill', color_discrete_sequence=['orange'])
fig.show()
```

```python
fig = px.bar(tips.groupby('day')['tip'].mean() , color_discrete_sequence=['orange'])
fig.show()
```

```python
fig = px.scatter( tips ,  x = 'total_bill' , y = 'tip' , color_discrete_sequence=['orange'])
fig.show()
```

```python
fig = px.box(tips , x = 'total_bill' , y= 'day',color_discrete_sequence=['orange'])
fig.show()
```
