### Loading the Libraries and Dataset

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings 
warnings.filterwarnings("ignore")

df = pd.read_csv('IPL.csv')
df.head()
```

### Some basic question

1: Which team won the most matches?

```python
match_win = df['match_winner'].value_counts()
match_win

```

```python
match_win = df['match_winner'].value_counts()
sns.barplot(y=match_win.index,x=match_win.values , palette='viridis')
plt.title("Most Match Win by Team")
```

### 2 : Toss Decision Trend

```python
sns.countplot(x = df['toss_decision'], palette='rainbow')
plt.title("Toss Decision Trend")
```

### 3: Toss Winner vs Match Winner

```python
df[df['toss_winner'] == df['match_winner']]
```

```python
df[df['toss_winner'] == df['match_winner']].count()
```

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
persentage = (count*100)/df.shape[0]
persentage.round(2)
```

### How to team win? (Run vs Wickets)

```python
sns.countplot(x = df['won_by'])
plt.title("Won By")
```

### Key Player Performanceabs

1 : Most "player of the match " Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values , y = count.index , palette='mako')
plt.title("Top ten players with man of the match")
```

### 2 Top Scorers

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind= 'barh')
```

### 10 Best bowling figures

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype('Int64')
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending= False).head(10)
top_bowlers.plot(kind= 'barh')
```
