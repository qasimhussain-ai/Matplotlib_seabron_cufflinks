Boxplot and Violinplot

boxplot and violinplot are used to show the distribution of categorical data . A box plot ( or box and whisker plot) shows the distribution 
of quantitative data in a way that facilitaties comparisions between variables or across levels of a categorical variable. The box shows the 
quartilies of the dataset while the whisker extend to show the rest of the distribution , except for points that are determined to be "outliers"
using a method that is a function of the 

```python
sns.violinplot(x = 'day', y = 'total_bill' , data=df)
```

Stripplot and swarmplot:
the stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn its own , but it is also a good complement
to a box or violin plot in cases where you want to show all observation along with some representation of the underlying distribution.

the swarmplot is similar to stripplot() , but the point are adjusted (along the categorical axis) so that they dont overlap . This gives a better 
representation of the distribution of values , although it does not scale as well to large numbers of observations (both in term of the ability
to show all the points and in terms of the computation needed to arrange them) 

```python
sns.stripplot(x = 'day' , y='total_bill' , data = df)
```

```python
sns.swarmplot(x = 'day' , y = 'total_bill' , data = df)
```

```python
sns.violinplot( x= 'tip', y = 'day' ,data=df)
sns.swarmplot(x = 'tip', y= 'day' , data = df)
```
